```python
# Add the libraries
import pandas as pd
import re
import os
import numpy as np
```

```python
elo_ratings = pd.read_csv('D:/STUDY/STUDY/PYTHON/Predict winner/elo rating.csv')
elo_ratings
```

```python
matches = pd.read_csv('D:/STUDY/STUDY/PYTHON/Predict winner/matches.csv')
matches
```

```python
team_data = pd.merge(team_stats, elo_ratings, left_on = 'Teams', right_on='Teams', how='inner').reset_index(drop=True)
team_data
```

```python
matches = pd.merge(matches, team_data, left_on = 'HOME_TEAM', right_on = 'Teams', how='inner').reset_index(drop=True)
matches.rename(columns = {'Attacks':'home_attack','Defense':'home_defense','elo rating':'home_elo_rating'}, inplace=True)
matches.drop(['Teams'], axis=1, inplace=True)
matches.drop(['Winner'], axis=1, inplace=True)
matches.head()
```

```python
matches = pd.merge(matches, team_data, left_on = 'AWAY_TEAM', right_on = 'Teams', how='inner').reset_index(drop=True)
matches.rename(columns = {'Attacks':'away_attack','Defense':'away_defense','elo rating':'away_elo_rating'}, inplace=True)
matches.drop(['Teams'], axis=1, inplace=True)
```

```python
matches['result'] = np.where(matches['HOME_TEAM_SCORE'] > matches['AWAY_TEAM_SCORE'], 'W',
                    np.where(matches['HOME_TEAM_SCORE'] < matches['AWAY_TEAM_SCORE'], 'L', 'D'))
```

```python
from sklearn.cluster import KMeans
cluster_data = team_data[['Attacks','Defense','elo rating']]
cluster_data['Attacks'] = cluster_data['Attacks'].astype('float')
cluster_data['Defense'] = cluster_data['Defense'].astype('float')
cluster_data['elo rating'] = cluster_data['elo rating'].astype('float')

cluster_data = (cluster_data-cluster_data.min())/(cluster_data.max()-cluster_data.min())
```

```python
# Clasify team into 'Weak', 'Average' and 'Strong' categories
kmeans = KMeans(n_clusters=3)
kmeans.fit(cluster_data)
cluster_data['Cluster'] = kmeans.labels_
cluster_numbers = cluster_data.groupby(['Cluster']).mean('elo rating').sort_values('elo rating')
cluster_order = cluster_numbers.index.unique()

team_data['Cluster']=cluster_data['Cluster'].astype('str')
team_data['Cluster'].replace(str(cluster_order[0]),'Weak', inplace=True)
team_data['Cluster'].replace(str(cluster_order[1]),'Average', inplace=True)
team_data['Cluster'].replace(str(cluster_order[2]),'Strong', inplace=True)
print(team_data.to_string())
```

```python
#the dict() function is called with team_data[['Teams','Cluster']].to_records(index=False) as an argument to create a dictionary 
# for mapping team names to their respective clusters, 
# then the resulting dictionary is used to map the team names in the HOME_TEAM and AWAY_TEAM columns to their corresponding clusters.

matches['home_cluster'] = matches['HOME_TEAM'].map(dict(team_data[['Teams','Cluster']].to_records(index=False)))
matches['away_cluster'] = matches['AWAY_TEAM'].map(dict(team_data[['Teams','Cluster']].to_records(index=False)))
```

```python
data = matches.dropna()
```

```python
# Since sklearn work with numeric data -> convert categories into numeric categories
data['home_cluster'].replace('Strong',3,inplace=True)
data['home_cluster'].replace('Average',2,inplace=True)
data['home_cluster'].replace('Weak',1,inplace=True)
data['away_cluster'].replace('Strong',3,inplace=True)
data['away_cluster'].replace('Average',2,inplace=True)
data['away_cluster'].replace('Weak',1,inplace=True)
data['result'].replace('W',3,inplace=True)
data['result'].replace('D',2,inplace=True)
data['result'].replace('L',1,inplace=True)
data
```

```python
data = data[['HOME_TEAM_SCORE', 'AWAY_TEAM_SCORE', 'result',
             'home_attack', 'home_defense', 'home_elo_rating', 'home_cluster',
             'away_attack', 'away_defense', 'away_elo_rating', 'away_cluster']]

data = data.apply(pd.to_numeric)
data['attack_effective'] = data['home_attack']-data['away_defense']
data['defense_effective'] = data['home_defense']-data['away_attack']
data['elo_rating_diff'] = data['home_elo_rating']-data['away_elo_rating']
data['team_strength_diff'] = data['home_cluster']-data['away_cluster']

data
```

```python
from sklearn.model_selection import train_test_split
X = data[['home_attack','home_defense', 'home_elo_rating', 'home_cluster',
          'away_attack','away_defense', 'away_elo_rating', 'away_cluster', 
          'attack_effective','defense_effective', 'elo_rating_diff','team_strength_diff']]

X = (X-X.min())/(X.max()-X.min())
y = data[['HOME_TEAM_SCORE', 'AWAY_TEAM_SCORE', 'result']]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, shuffle=False)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=5)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test['result'], y_pred[:,2])
print("Accuracy: ",accuracy)
     
```
